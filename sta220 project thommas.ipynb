```python
import re
import time
from bs4 import BeautifulSoup
import requests

# Function to scrape data from a single URL
def scrape_data(url):
    try:
        # Get the HTML content of the page
        parent = requests.get(url)
        html = parent.text

        # Parse the HTML using BeautifulSoup
        soup = BeautifulSoup(html, "html.parser")

        # Find the establishment name
        establishment_name_element = soup.find(class_="establishment-page-left-name")
        if establishment_name_element:
            establishment_name = establishment_name_element.get_text(strip=True)
        else:
            raise ValueError("Establishment name not found")

        # Find the div containing the establishment address
        address_div = establishment_name_element.find_next("div")
        if address_div:
            establishment_address = address_div.get_text(strip=True)
        else:
            raise ValueError("Establishment address not found")

        # Find the div containing the inspection date
        inspection_date_div = soup.find(class_="w-col w-col-8")
        if inspection_date_div:
            inspection_date_text = inspection_date_div.get_text(strip=True)
            inspection_date = inspection_date_text.split("|")[0].strip()
        else:
            raise ValueError("Inspection date not found")

        # Find all elements with class="observations-text"
        observations = soup.find_all(class_="observations-text")

        # Define a regular expression pattern to match immediate numbers or numbers followed by a single letter prior to the colon
        pattern = r"\b\d+[a-z]?(?=: )"
        # Initialize a list to store the extracted codes
        extracted_codes = []

        # Iterate over each observation
        for observation in observations:
            # Find all matches of the pattern within the observation text
            matches = re.findall(pattern, observation.text)
            # Add the matches to the list of extracted codes
            extracted_codes.extend(matches)

        # Construct the result list
        result = [establishment_name, establishment_address, extracted_codes, inspection_date, url]

        return result

    except Exception as e:
        print(f"Error scraping {url}: {e}")
        return None

# List to store results
results = []


# Loop through the list of URLs
for url in links2[5000:5696]:
    # Scrape data from the current URL
    result = scrape_data(url)
    if result:
        # Append the result to the list of results
        results.append(result)
    
    # Pause for 1 second before making the next request
    time.sleep(0.25)

# Print the collected results
for result in results:
    print(result)
```

```python
import json

# Save the results as JSON
with open("results_chunk_10.json", "w") as json_file:
    json.dump(results, json_file)

print("Results saved to results_chunk_10.json")
```
